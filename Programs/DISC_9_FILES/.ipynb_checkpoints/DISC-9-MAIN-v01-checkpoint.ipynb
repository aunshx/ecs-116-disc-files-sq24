##### <span style=color:blue>Aggregation Queries</span>

```python
# Calculate the total number of employees per department

pipeline = [
    {"$group": {"_id": "$dno", "count": {"$sum": 1}}},
    {"$lookup": {
        "from": "departments",
        "localField": "_id",
        "foreignField": "dnumber",
        "as": "department_info"
    }},
    {"$unwind": "$department_info"},
    {"$project": {"department_name": "$department_info.dname", "count": 1}},
    {"$sort": {"count": -1}}
]

# Execute the aggregation pipeline
total_employees_per_department = db.employees.aggregate(pipeline)

# Print the results
pp.pprint(list(total_employees_per_department))


# The above query will look like follows in SQL

# The table name employees is employee_with_comments as that's how I've imported
# it in psql

# SELECT d.dname AS department_name, COUNT(e.ssn) AS count
# FROM employee_with_comments e
# JOIN department d ON e.dno = d.dnumber
# GROUP BY d.dname
# ORDER BY count DESC;
```

```python
# Average Salary by Department

avg_salary_by_dept = db.employees.aggregate([
    {
        # Group employees by department number and calculate the average salary
        "$group": {
            "_id": "$dno",
            "averageSalary": { "$avg": "$salary" }
        }
    },
    {
        # Join the grouped results with the departments collection to get department names
        "$lookup": {
            "from": "departments",
            "localField": "_id",
            "foreignField": "dnumber",
            "as": "department"
        }
    },
    {
        # Unwind the joined department array to convert it to a single object
        "$unwind": "$department"
    },
    {
        # Project the department name and average salary
        "$project": {
            "_id": 0,
            "departmentName": "$department.dname",
            "averageSalary": 1
        }
    }
])

for dept in avg_salary_by_dept:
    pp.pprint(dept)
    
# The above query will look like follows in SQL

# The table name employees is employee_with_comments as that's how I've imported
# it in psql

# SELECT d.dname AS departmentName, AVG(e.salary) AS averageSalary
# FROM employee_with_comments e
# JOIN department d ON e.dno = d.dnumber
# GROUP BY d.dname;
```

```python
# Average Project Hours by Employee with Comments
avg_project_hrs_w_comments = db.works_on.aggregate([
    {
        # Join the works_on collection with the employees collection on SSN
        "$lookup": {
            "from": "employees",
            "localField": "essn",
            "foreignField": "ssn",
            "as": "employee"
        }
    },
    {
        # Unwind the joined employee array to convert it to a single object
        "$unwind": "$employee"
    },
    {
        # Group by employee SSN and calculate the average hours worked, retaining the first comment for each employee
        "$group": {
            "_id": "$essn",
            "avgHours": { "$avg": "$hours" },
            "comments": { "$first": "$employee.comments" }
        }
    },
    {
        # Join the grouped results with the employees collection to get employee details
        "$lookup": {
            "from": "employees",
            "localField": "_id",
            "foreignField": "ssn",
            "as": "employee"
        }
    },
    {
        # Unwind the joined employee array to convert it to a single object
        "$unwind": "$employee"
    },
    {
        # Project the employee's first name, last name, average hours worked, and comments
        "$project": {
            "_id": 0,
            "fname": "$employee.fname",
            "lname": "$employee.lname",
            "avgHours": 1,
            "comments": 1
        }
    }
])

for dept in avg_project_hrs_w_comments:
    pp.pprint(dept)
```

```python
# Average Salary by Department with Employee Count

# ATTENTION - DO NOT RUN THIS!!!!!!

# The table name employees is employee_with_comments as that's how I've imported
# it in psql

SELECT d.dname AS departmentName, AVG(e.salary) AS averageSalary, COUNT(e.ssn) AS employeeCount
FROM employee_with_comments e
JOIN department d ON e.dno = d.dnumber
GROUP BY d.dname;
```

```python
# Mongo Equivalent of above SQL query

avg_sal_dept_emp_count = db.employees.aggregate([
    {
        # Group employees by department number and calculate the average salary and count of employees
        "$group": {
            "_id": "$dno",
            "averageSalary": { "$avg": "$salary" },
            "employeeCount": { "$sum": 1 }
        }
    },
    {
        # Join the grouped results with the departments collection to get department names
        "$lookup": {
            "from": "departments",
            "localField": "_id",
            "foreignField": "dnumber",
            "as": "department"
        }
    },
    {
        # Unwind the joined department array to convert it to a single object
        "$unwind": "$department"
    },
    {
        # Project the department name, average salary, and employee count
        "$project": {
            "_id": 0,
            "departmentName": "$department.dname",
            "averageSalary": 1,
            "employeeCount": 1
        }
    }
])

for dept in avg_sal_dept_emp_count:
    pp.pprint(dept)
```

```python
# Total Hours Worked by Each Employee on Each Project

# ATTENTION - DO NOT RUN THIS!!!!!!

# The table name employees is employee_with_comments as that's how I've imported
# it in psql

SELECT CONCAT(e.fname, ' ', e.lname) AS employeeName, p.pname AS projectName, SUM(w.hours) AS totalHours
FROM works_on w
JOIN employee_with_comments e ON w.essn = e.ssn
JOIN project p ON w.pno = p.pnumber
GROUP BY e.fname, e.lname, p.pname;
```

```python
# Mongo Equivalent of above SQL query

hrs_worked_per_proj_per_emp = db.works_on.aggregate([
    {
        # Group by employee SSN and project number and calculate the total hours worked
        "$group": {
            "_id": {
                "essn": "$essn",
                "pno": "$pno"
            },
            "totalHours": { "$sum": "$hours" }
        }
    },
    {
        # Join the grouped results with the employees collection to get employee names
        "$lookup": {
            "from": "employees",
            "localField": "_id.essn",
            "foreignField": "ssn",
            "as": "employee"
        }
    },
    {
        # Unwind the joined employee array to convert it to a single object
        "$unwind": "$employee"
    },
    {
        # Join the grouped results with the projects collection to get project names
        "$lookup": {
            "from": "projects",
            "localField": "_id.pno",
            "foreignField": "pnumber",
            "as": "project"
        }
    },
    {
        # Unwind the joined project array to convert it to a single object
        "$unwind": "$project"
    },
    {
        # Project the employee name, project name, and total hours worked
        "$project": {
            "_id": 0,
            "employeeName": { "$concat": ["$employee.fname", " ", "$employee.lname"] },
            "projectName": "$project.pname",
            "totalHours": 1
        }
    }
])

for dept in hrs_worked_per_proj_per_emp:
    pp.pprint(dept)
```

```python
# Find Employees with Positive Comments

# NOTE: You will get an error here as text search is indexed in MongoDB and it hasn't been set yet

positive_comments = list(db.employees.find(
    {"$text": {"$search": "good nice amazing"}}
))

print("Positive Comments:")
for comment in positive_comments:
    print(comment)
```

```python
# Create Text Index
db.employees.create_index([("comments", "text")])
```

```python
# Run the query again 

positive_comments = list(db.employees.find(
    {"$text": {"$search": "good nice amazing"}}
))

print("Positive Comments:")
for comment in positive_comments:
    pp.pprint(comment)
```

```python
# What if I only want the employee name and the comment?

positive_comments = list(db.employees.find(
    {"$text": {"$search": "good nice amazing"}},
    {"fname": 1, "lname": 1, "comments": 1, "_id": 0}
))

# Store the count of positive comments
positive_comments_count = len(positive_comments)

print('Positive Comment Count', positive_comments_count)

print("Positive Comments:")
for comment in positive_comments:
    print(comment)
    
# The above query will look like follows in SQL
    
# SELECT fname, lname, comments
# FROM employee_with_comments
# WHERE comments LIKE '%good%' OR comments LIKE '%nice%' OR comments LIKE '%amazing%';
```

```python
# Find Employees with Negative Comments

# We will use the keywords [bad, worse]

negative_comments = list(db.employees.find(
    {"$text": {"$search": "bad worse"}},
    {"fname": 1, "lname": 1, "comments": 1, "_id": 0}
))

# Store the count of negative comments
negative_comments_count = len(negative_comments)

print('Negative Comment Count', negative_comments_count)

print("\nNegative Comments:")
for comment in negative_comments:
    print(comment)
    

# The above query will look like follows in SQL
    
# SELECT fname, lname, comments
# FROM employee_with_comments
# WHERE comments LIKE '%bad%' OR comments LIKE '%worse%';
```
